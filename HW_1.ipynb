# Задача 1
Группа студентов изучает 10 различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть 4 разных занятия?

Решение.
Важен состав и порядот

```python
def fact(n):
    if(n==1 or n==0):
        return 1 
    else:
        return n*fact(n-1)
n=10
m=4
a=fact(n)/fact(n-m)
print(a)
```

# Задача 2
Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

```python
# Решение

# Найдем число сочетаний 1 туз из 4
b=fact(4)/(fact(1)*fact(4-1))
# выбор 3-х карт из колоды без тузов 
c=fact(32)/(fact(3)*fact(32-3))
# Число сочетаний
d=b*c
print(b,c,d)
```

```python
# Решение
# Всего 6 пар на скамейке. Остальные могут сесть 5! Всего вариантов сесть 2м рядом = 6*2*5!

n=7
a=6*2*fact(5)/fact(7)
print("Вероятность: "a)
```

# Задача 4

Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

```python
# Решение А
# Отпределить варианты, что из 50 карт были выбраны 3. Порядок не важен.
n=50
m=3
a=fact(n)/(fact(m)*fact(n-m))
# Варианты выбора 3 карт из 60
c=fact(60)/(fact(3)*fact(60-3))
d=a/c
print(a,c, "Вероятность: ",d)
```

```python
# Решение B
# Отпределить варианты, что из 50 карт были выбраны 2. Порядок не важен.
n=50
m=2
a=fact(n)/(fact(m)*fact(n-m))
# Варианты выбора 1 карт из оставшихся 10. Порядок не важен
b=fact(10)/(fact(1)*fact(10-1))
# Варианты выбора 3 карт из 60
c=fact(60)/(fact(3)*fact(60-3))
d=a*b/c
print(a,b,c, "Вероятность:",d)
```

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Вероятность выпадения четного числа: 3/6=1/3

Вероятность выпадения числа больше 3-х: 3/6=1/3

Вероятность появления обоих событий P(B|A): 1/3*1/3=1/9. Так как 1/9<>1/3, то события являются независимыми.
