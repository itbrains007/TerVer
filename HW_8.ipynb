## Задача 1

Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: 

football_players = [173, 175, 180, 178, 177, 185, 183, 182]

hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]

lifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]

Будем проверят ьгипотещу что все равны.

```python
import numpy as np
from scipy import stats
```

```python
football_players = np.array([173, 175, 180, 178, 177, 185, 183, 182], dtype=np.float64)

hockey_players = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180], dtype=np.float64)

lifters = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170], dtype=np.float64)

football_n = football_players.shape[0]
hockey_n = hockey_players.shape[0]
lifters_n = lifters.shape[0]
```

Найдем серднее:

```python
football_mean = football_players.mean()
hockey_mean = hockey_players.mean()
lifters_mean = lifters.mean()

football_mean, hockey_mean, lifters_mean
```

Видно, что средний рост разный. Проверим статистическую значимость этого отличия.

Соберём все значения в один массив:

```python
all = np.concatenate([football_players, hockey_players, lifters])

print(all)
```

```python
all_mean = all.mean()
all_mean
```

Найдём значения $S_b$ и $S_{w}$:

```python
S2_b = football_n * (football_mean - all_mean) ** 2 + hockey_n * (hockey_mean - all_mean) ** 2 + lifters_n * (lifters_mean - all_mean) ** 2

S2_w = ((football_players - football_mean) ** 2).sum() + ((hockey_players - hockey_mean) ** 2).sum() + ((lifters - lifters_mean) ** 2).sum()

S2_b, S2_w
```

Проверим выполнение равенства $S^2 = S_b^2 + S_{w}^2:$

```python
S2_b + S2_w - ((all - all_mean) ** 2).sum()
#S2_b + S2_w, ((all - all_mean) ** 2).sum()
```

Считаем равны

Запишем оценки дисперсий:

```python
k = 3
n = football_n + hockey_n + lifters_n

k1 = k - 1
k2 = n - k

sigma2_b = S2_b / k1
sigma2_w = S2_w / k2

sigma2_b, sigma2_w
```

```python
T = sigma2_b / sigma2_w
T
```

```python
alpha = 0.05

F_crit = stats.f.ppf(1 - alpha, k1, k2)
F_crit
```

```python
T-F_crit
```

Видим, что $T > F_{crit}$, поэтому заключаем, что отличие среднего роста действительно является статистически значимым.

```python
p = 1 - stats.f.cdf(T, k1, k2)
p
```
